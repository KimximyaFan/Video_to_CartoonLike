```python
import cv2 as cv
import numpy as np

# 프로그램 텍스트 안내

canvas = np.full( (480, 640, 3), 255, dtype=np.uint8 )
cv.putText(canvas, " press '1' or '2' for video type ", (200, 240), cv.FONT_HERSHEY_DUPLEX, 0.5, (0, 0, 0) )
cv.imshow('Video_to_Cartoon', canvas)

# 1번 누르면 1번 영상, 2번 누르면 2번 영상을 재생

while(True):
    key = cv.waitKey(0)
    
    if key == ord('1') or key == ord('2'):
        break

cv.destroyAllWindows()
        
if key == ord('1'):        
    video_file = 'Video_1.mp4'
else:
    video_file = 'Video_2.mp4'

video = cv.VideoCapture(video_file)

if video.isOpened() == False:
    exit()

fps = 50
wait_msec = int( 1000 / fps )

# Bilateral Filter Variableds
kernel_size = 9
sigma_color = 150
sigma_space = 2.4

# ===================================================================

def Cartoon_Like( img ):
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    gray = cv.medianBlur(gray, 5)
    edges = cv.adaptiveThreshold(gray, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 9, 9)
    img_tran = cv.bilateralFilter(img, kernel_size, sigma_color, sigma_space)
    img_tran = cv.bitwise_and(img_tran, img_tran, mask=edges)
    return img_tran
    
while True:
    valid, img = video.read()
    
    if not valid:
        break
    
    cv.imshow('Video Player', Cartoon_Like(img))
    key = cv.waitKey(wait_msec)
    
    if key == 27:
        break

        
video.release()
cv.destroyAllWindows()
```
